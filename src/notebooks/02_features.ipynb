```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('../data/sample_data.csv')
print('Original shape:', df.shape)
df.head()
```

```python
# Feature 1 — Tenure Groups
df['tenure_group'] = pd.cut(df['tenure'],
                             bins=[0, 12, 24, 48, 72],
                             labels=['0-1yr', '1-2yr', '2-4yr', '4-6yr'])
print('Tenure group distribution:')
print(df['tenure_group'].value_counts())
```

```python
# Feature 2 — Charge Ratio
df['charge_ratio'] = df['MonthlyCharges'] / (df['TotalCharges'] + 1)
print('Charge ratio stats:')
print(df['charge_ratio'].describe())
```

```python
# Feature 3 — High Risk Flag
df['high_risk'] = ((df['NumSupportTickets'] >= 3) &
                    (df['Contract'] == 'Month-to-month')).astype(int)
print('High risk customers:')
print(df['high_risk'].value_counts())
```

```python
# Visualize new features vs Churn
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

sns.countplot(data=df, x='tenure_group', hue='Churn',
              palette=['#2ecc71','#e74c3c'], ax=axes[0])
axes[0].set_title('Tenure Group vs Churn')

sns.boxplot(data=df, x='Churn', y='charge_ratio',
            palette=['#2ecc71','#e74c3c'], ax=axes[1])
axes[1].set_title('Charge Ratio vs Churn')

sns.countplot(data=df, x='high_risk', hue='Churn',
              palette=['#2ecc71','#e74c3c'], ax=axes[2])
axes[2].set_title('High Risk Flag vs Churn')

plt.tight_layout()
plt.show()
```

```python
# Save engineered dataset
df.to_csv('../data/engineered_data.csv', index=False)
print('✅ Engineered dataset saved!')
print('New shape:', df.shape)
print('New features:', ['tenure_group', 'charge_ratio', 'high_risk'])
```

## 🔑 Feature Engineering Summary
| Feature | Type | Business Meaning |
|---------|------|------------------|
| tenure_group | Categorical | Customer lifecycle stage |
| charge_ratio | Numerical | Monthly vs total spend ratio |
| high_risk | Binary | Combined risk flag |
